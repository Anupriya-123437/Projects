```python
#classify movies based on ratings[Excellent, good and average]
data.columns
def rating(IMDB_Rating):
    if IMDB_Rating>=7.0:
        return "Excellent"
    elif IMDB_Rating >=6.0:
        return "Good"
    else:
        return "Average"

data['Rating_cat'] = data['IMDB_Rating'].apply(rating)
data.head()
```

```python
#find unique values from genre
#1. create one list for all items present in genre column
list1 = []
for value in data['Genre']:
    list1.append(value.split(','))

list1
```

```python
#2. convert this 2d list into 1d list
one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)

one_d
```

```python
#3. find unique values available in one_d
unique_list = []
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)

unique_list
```
